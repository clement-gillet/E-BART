### 2) Visualization

A random data point is picked and its structure is outlined

```python
print("---------------------------------------------------------------------")
print(train_maven_ere.iloc[675])
print("---------------------------------------------------------------------")
print('\n','Document (in sentences): ', '\n')
print(train_maven_ere.iloc[675].sentences, '\n')
print("Events: ", '\n')
[print(train_maven_ere.iloc[675].events[i], '\n') for i in range(len(train_maven_ere.iloc[675].events))]
print("TIMEX (when the text mentions a specific DATE/TIME/DURATION/PREPOSTEX): ", '\n')
[print(train_maven_ere.iloc[675].TIMEX[i], '\n') for i in range(len(train_maven_ere.iloc[675].TIMEX))]
```

```python
print("Temporal relations: ", '\n')
# Temporal relations keys : ['BEFORE', 'OVERLAP', 'CONTAINS', 'SIMULTANEOUS', 'ENDS-ON', 'BEGINS-ON']
print("BEFORE: ", '\n')
[print(train_maven_ere.iloc[675].temporal_relations['BEFORE'][i], '\n') for i in range(len(train_maven_ere.iloc[675].temporal_relations['BEFORE']))]
print("OVERLAP: ", '\n')
[print(train_maven_ere.iloc[675].temporal_relations['OVERLAP'][i], '\n') for i in range(len(train_maven_ere.iloc[675].temporal_relations['OVERLAP']))]
print("CONTAINS: ", '\n')
[print(train_maven_ere.iloc[675].temporal_relations['CONTAINS'][i], '\n') for i in range(len(train_maven_ere.iloc[675].temporal_relations['CONTAINS']))]
print("SIMULTANEOUS: ", '\n')
[print(train_maven_ere.iloc[675].temporal_relations['SIMULTANEOUS'][i], '\n') for i in range(len(train_maven_ere.iloc[675].temporal_relations['SIMULTANEOUS']))]
print("ENDS-ON: ", '\n')
[print(train_maven_ere.iloc[675].temporal_relations['ENDS-ON'][i], '\n') for i in range(len(train_maven_ere.iloc[675].temporal_relations['ENDS-ON']))]
print("BEGINS-ON: ", '\n')
[print(train_maven_ere.iloc[675].temporal_relations['BEGINS-ON'][i], '\n') for i in range(len(train_maven_ere.iloc[675].temporal_relations['BEGINS-ON']))]
```

```python

print("Causal relations: ", '\n')
# Temporal relations keys : ['CAUSE', 'PRECONDITION']
print('CAUSE:', '\n')
[print(train_maven_ere.iloc[675].causal_relations['CAUSE'][i], '\n') for i in range(len(train_maven_ere.iloc[675].causal_relations['CAUSE']))]
print('PRECONDITION:', '\n')
[print(train_maven_ere.iloc[675].causal_relations['PRECONDITION'][i], '\n') for i in range(len(train_maven_ere.iloc[675].causal_relations['PRECONDITION']))]
```

```python
# Is the Subevent relations array always empty??
print("Subevent relations: ", '\n')
[print(train_maven_ere.iloc[675].subevent_relations, '\n') for i in range(len(train_maven_ere.iloc[675].subevent_relations))]
```
